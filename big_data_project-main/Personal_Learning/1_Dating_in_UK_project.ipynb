```python
# W5 REMOVE EMOJIS
import re
def remove_emojis(data):
    emoj = re.compile("["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
        u"\U00002500-\U00002BEF"  # chinese char
        u"\U00002702-\U000027B0"
        u"\U00002702-\U000027B0"
        u"\U000024C2-\U0001F251"
        u"\U0001f926-\U0001f937"
        u"\U00010000-\U0010ffff"
        u"\u2640-\u2642" 
        u"\u2600-\u2B55"
        u"\u200d"
        u"\u23cf"
        u"\u23e9"
        u"\u231a"
        u"\ufe0f"  # dingbats
        u"\u3030"
                      "]+", re.UNICODE)
    return re.sub(emoj, '', data)
```

```python
w['full_text'] = w['full_text'].apply(remove_emojis)
```

```python
#Tokenise data :: split words 
w['tokens'] = w['full_text'].apply(TweetTokenizer().tokenize)
```

```python
#remove stop words
stopwords_vocabulary = stopwords.words('english')

w['stopwords'] = w['tokens'].apply(lambda x: [i for i in x if i.lower() not in stopwords_vocabulary])
```

```python
#remove punctuation 
punctuations = list(string.punctuation)
w['punctuation'] = w['stopwords'].apply(lambda x: [i for i in x if i not in punctuations])
```

```python
########################################################
#Sentiment Analysis
```

```python
w['final'] = w['punctuation'].apply(lambda x: [i for i in x if len(i) > 1])
```

```python
sentiment = SentimentIntensityAnalyzer()
w['sentiment'] = w['final'].apply(lambda x: sentiment.polarity_scores(' '.join(x))['compound'])
```

```python
w
```

```python
feelings =[]
for i in w["sentiment"]:
    feelings.append(i)
```

```python
pos = len(w[w.sentiment > 0])
neg = len(w[w.sentiment < 0])
neu = len(w[w.sentiment == 0])

print(pos, neg, neu)
```

```python
#find average of numbers 
import statistics
statistics.mean(feelings)
```
